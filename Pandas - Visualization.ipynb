```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\My Folder\Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
# DIFFERENT STYLES OF VISUALIZATONS
print(plt.style.available)
plt.style.use('dark_background')
```

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
# SPECIFYING COLUMN/VALUES
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
# HORIZONTAL STACKED BAR CHART
df.plot.barh(stacked = True)
```

```python
# SCATTER PLOT
```

```python
# HAS TO SPECIFY X & Y AXES
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300, c = 'darkred')
```

```python
# HISTOGRAM
```

```python
# SPECIFY BINS, REPRESENT DISTRIBUTION IN THE X-AXIS
df.plot.hist(bins = 20)
```

```python
# BOX PLOT (SHOWS MIN, 25%, 50%, 75%, MAX)
```

```python
df.boxplot()
```

```python
# AREA PLOT
```

```python
df.plot.area( figsize = (10,4) )
```
