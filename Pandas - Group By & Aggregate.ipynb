# Group By & Aggregating

```python
# GROUP BY - GROUP TOGETHER THE VALUES IN A COLUMN, DISPLAYING THEM ON THE SAME ROW
# ENABLING USER TO PERFORM AGGREGATE FUNCTIONS ON THOSE GROUPINGS
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\My Folder\Pandas\Flavors.csv")
df
```

```python
# GROUP BY
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
# AGGREGATING
```

```python
# AGGREGATION ONLY TAKES COLUMNS W/ INT
group_by_frame.mean(numeric_only = True)
```

```python
# COMBINATION IN ONE LINE
df.groupby('Base Flavor').mean(numeric_only = True)
```

```python
# COUNTING BASED ON 'BASE FLAVOR' COLUMN
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum(numeric_only = True)
```

```python
# AGGREGATION FUNCTION - agg( {dict} )
```

```python
df.groupby('Base Flavor').agg( {'Flavor Rating':['mean','max','count','sum'], 'Texture Rating':['mean','max','count','sum']} )
```

```python
df
```

```python
# GROUP BY MULTIPLE COLUMNS
```

```python
df.groupby( ['Base Flavor','Liked'] ).mean(numeric_only = True)
```

```python
df.groupby( ['Base Flavor','Liked'] ).agg( {'Flavor Rating':['mean','max','count','sum']} )
```
