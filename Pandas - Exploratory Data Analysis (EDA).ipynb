# Exploratory Data Analysis (EDA)

```python
# EDA IS BASICALLY JUST THE FIRST LOOK AT DATA
# IDENTIFYING PATTERNS WITHIN DATA
# UNDERSTAND THE RELATIONSHIP BETWEEN FEATURES
# LOOKING AT OUTLIER(S) THAT MAY EXIST(S) WITHIN DATASET
# LOOKING FOR MISTAKES / MISSING VALUES THAT NEEDS TO BE CLEANED
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\My Folder\Pandas\world_population_EDA.csv")
df
```

```python
# CHANGE DECIMAL POINTS
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# HIGH-LEVEL INFORMATION
df.info()
```

```python
# DESCRIBE() - HIGH LEVEL OVERVIEW
df.describe()
```

```python
# COUNT ALL NULL VALUES
df.isnull().sum()
```

```python
# COUNT UNIQUE VALUES IN EACH COLUMN
df.nunique()
```

```python
# TOP 10 LARGEST BY POPULATION
df.sort_values(by = 'World Population Percentage', ascending = False).head(10)
```

```python
# CORELATION BETWEEN NUMERIC VALUES (USUALLY)
df.corr()
```

```python
# VISUALIZE USING SEABORN VIA HEATMAP
sns.heatmap(df.corr(), annot = True)
plt.rcParams['figure.figsize'] = (18,7)
plt.show()
```

```python
# GROUPING DATA
# USE CASE: ARE THERE CERTAIN CONTINENTS THAT GROW FASTER? IF SO, IN WHAT WAYS?
df.groupby('Continent').mean().sort_values(by = '2022 Population', ascending = False)
# AVERAGE POPULATION OF EACH CONTIENENT, SORT FROM H -> l BASED ON 2022 POPULATION
```

```python
# LOOKING AT OCEANIA COUNTRIES
df[df['Continent'].str.contains('Oceania')]
```

```python
# VISUALIZE, SELECTING CERTAIN COLUMNS
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by = '2022 Population', ascending = False)
df2
```

```python
# LIST AND COPY COLUMNS, DO NOT NEED TO TYPE ONE BY ONE
df.columns
```

```python
# SWITCHING THE 0 & 1 AXES
df3 = df2.transpose()
df3
```

```python
# PLOT LINE GRAPH
df3.plot()
```

```python
# BOX PLOT - GREAT FOR FINDING OUTLIERS (CLEANNING DATA) !
# SHOWS QUARTILES, UPPER, LOWER, OUTLIERS
```

```python
# CIRCLES REPRESENTS OUTLIER VALUES, EACH VALUE REPRESENT A COUNTRY
df.boxplot()
```
