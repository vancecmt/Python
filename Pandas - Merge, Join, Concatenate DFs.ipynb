# Merge, Join, Concatenate

```python
# COMBINE TWO SEPARATE DFs INTO ONE 
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\My Folder\Pandas\LOTR 1.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\My Folder\Pandas\LOTR 2.csv")
df2
```

```python
# MERGE (SIMILAR TO JOINS)
```

```python
# DF1 = LEFT, DF2 = RIGHT
# INNER JOIN
#df1.merge(df2)
df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
# OUTER JOIN
# NaN = NOT A NUMBER
df1.merge(df2, how = 'outer')
```

```python
# LEFT JOIN
df1.merge(df2, how = 'left')
```

```python
# RIGHT JOIN
df1.merge(df2, how = 'right')
```

```python
# CROSS JOIN
# TAKES EACH VALUE FROM l DF AND COMPARES TO EACH VALUES FROM R DF
df1.merge(df2, how = 'cross')
```

```python
# JOIN
```

```python
# JOIN MIST SPECIFY SUFFIX BECAUSE NO DEFAULT LIKE MERGE
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_L', rsuffix = '_R')
```

```python
# USUALLY WORK BETTER WITH INDEXES
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_L', rsuffix = '_R', how = 'outer')
df3
```

```python
# CONCATENATE
```

```python
# PUTTING ONE DF ON TOP OF ANOTHER, INSTEAD OF NEXT TO ANOTHER (MERGE/JOIN)
```

```python
# NOTICE 1001 & 1OO2, AGE ARE NOT MERGED/JOUINED (ON TOP, INSTEAD OF NEXT)
pd.concat( [df1,df2] )
```

```python
pd.concat( [df1,df2], join = 'inner' )
```

```python
pd.concat( [df1,df2], join = 'outer' )
```

```python
pd.concat( [df1,df2], join = 'outer', axis = 1 )
```

```python
# APPEND
# APPEND ROWS FROM ONE DF TO THE END OF ANOTHER DF, AND RETURN THAT DF
# NOT RECOMMENDED!!
```

```python
df1.append(df2)
```
